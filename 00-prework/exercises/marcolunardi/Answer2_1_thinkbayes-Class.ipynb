Adding class 'Bowl' to implement selection without replacement:

```python
from thinkbayes import Pmf
from __future__ import division

class Bowl:
    
    def __init__(self, nv, nc):
        self.nv = nv
        self.nc = nc
        
    def update(self, data):
        if data == 'vanilla':
            self.nv = self.nv-1
        if data == 'chocolate':
            self.nc = self.nc-1

class Cookie(Pmf):
    """A map from string bowl ID to probablity."""

    def __init__(self, hypos):
        """Initialize self.
        hypos: sequence of string bowl IDs
        """
        Pmf.__init__(self)
        for hypo in hypos:
            self.Set(hypo, 1)
        self.Normalize()

    def Update(self, data, likev1, likec1, likev2, likec2):
        """Updates the PMF with new data.
        data: string cookie type
        """
        self.likev1 = likev1
        self.likec1 = likec1
        self.likev2 = likev2
        self.likec2 = likec2
        for hypo in self.Values():
            like = self.Likelihood(data, hypo)
            self.Mult(hypo, like)
        self.Normalize()

    def Likelihood(self, data, hypo):
        """The likelihood of the data under the hypothesis.
        data: string cookie type
        hypo: string bowl ID
        """
        self.mixes = {
            'Bowl 1':dict(vanilla=self.likev1, chocolate=self.likec1),
            'Bowl 2':dict(vanilla=self.likev2, chocolate=self.likec2),
            }
        mix = self.mixes[hypo]
        like = mix[data]
        return like

def main():
    
    Bowl1 = Bowl(30,10)
    Bowl2 = Bowl(20,20)

    likev1 = Bowl1.nv/(Bowl1.nv+Bowl1.nc)
    likec1 = Bowl1.nc/(Bowl1.nv+Bowl1.nc)
    likev2 = Bowl2.nv/(Bowl2.nv+Bowl2.nc)
    likec2 = Bowl2.nc/(Bowl2.nv+Bowl2.nc)
    
    hypos = ['Bowl 1', 'Bowl 2']
    
    pmf = Cookie(hypos)

    pmf.Update('vanilla', likev1, likec1, likev2, likec2)
    
    print 'Bowl probabilities for 1st extracted and eaten cookie (vanilla):'
    for hypo, prob in pmf.Items():
        print hypo, prob
        pmf.Set(hypo,prob)
        
    Bowl1.update('vanilla')
    Bowl2.update('vanilla')
    
    likev1 = Bowl1.nv/(Bowl1.nv+Bowl1.nc)
    likec1 = Bowl1.nc/(Bowl1.nv+Bowl1.nc)
    likev2 = Bowl2.nv/(Bowl2.nv+Bowl2.nc)
    likec2 = Bowl2.nc/(Bowl2.nv+Bowl2.nc)

    pmf.Update('chocolate', likev1, likec1, likev2, likec2)
    
    print 'Bowl probabilities for 2nd extracted and eaten cookie (chocolate):'
    for hypo, prob in pmf.Items():
        print hypo, prob
        pmf.Set(hypo,prob)
        
    Bowl1.update('chocolate')
    Bowl2.update('chocolate')
    
    likev1 = Bowl1.nv/(Bowl1.nv+Bowl1.nc)
    likec1 = Bowl1.nc/(Bowl1.nv+Bowl1.nc)
    likev2 = Bowl2.nv/(Bowl2.nv+Bowl2.nc)
    likec2 = Bowl2.nc/(Bowl2.nv+Bowl2.nc)

    pmf.Update('vanilla', likev1, likec1, likev2, likec2)
    
    print 'Bowl probabilities for 3rd extracted and eaten cookie (vanilla):'
    for hypo, prob in pmf.Items():
        print hypo, prob
        pmf.Set(hypo,prob)
        
    Bowl1.update('vanilla')
    Bowl2.update('vanilla')
        
if __name__ == '__main__':
    main()
```
