(Think Stats Ex 3.1)

Something like the class size paradox appears if you survey children and ask how many children are in their family. Families with many children are more likely to appear in your sample, and families with no children have no chance to be in the sample.

Use the NSFG respondent variable NUMKDHH to construct the actual distribution for the number of children under 18 in the household.

Now compute the biased distribution we would see if we surveyed the children and asked them how many children under 18 (including themselves) are in their household.

Plot the actual and biased distributions, and compute their means. As a starting place, you can use chap03ex.ipynb. This is an ipython notebook from the ThinkStats2 repository.

```python
import chap01soln
resp = chap01soln.ReadFemResp()
resp.columns.values
```

```python
import thinkstats2
ueighteen = resp['numkdhh']
actual_pmf = thinkstats2.Pmf(ueighteen, label='actual')
print(actual_pmf)
```

```python
biased_pmf = actual_pmf.Copy(label="child's perspective")
for x, p in actual_pmf.Items():
    biased_pmf.Mult(x,x)
biased_pmf.Normalize()
print(biased_pmf)
```

```python
import thinkplot
thinkplot.PrePlot(2)
thinkplot.Pmfs([actual_pmf, biased_pmf])
thinkplot.Show(xlabel='number of children', ylabel='probability')
```

```python
print('actual mean', actual_pmf.Mean())
print('biased mean', biased_pmf.Mean())
```

### <a id='3'></a> Exercise 3 
(Think Stats Ex 4.2)

The numbers generated by random.random are supposed to be uniform between 0 and 1; that is, every value in the range should have the same probability.
Generate 1000 numbers from random.random and plot their PMF and CDF. Is the distribution uniform?

```python
import random
random_numbers = [random.random() for _ in range(1000)]

import thinkstats2, thinkplot
pmf = thinkstats2.Hist(random_numbers)
thinkplot.Pmf(pmf)
thinkplot.Show(ylabel='probability',xlabel='value')

cdf = thinkstats2.Cdf(random_numbers, label='numbers')
thinkplot.Cdf(cdf)
thinkplot.Show(ylabel='CDF', xlabel='number')
```

####Answer:
Pretty darn uniform.

(Think Stats Ex 7.1)

Using data from the NSFG, make a scatter plot of birth weight versus mother’s age. Plot percentiles of birth weight versus mother’s age. Compute Pearson’s and Spearman’s correlations. How would you characterize the relationship between these variables?

```python
import nsfg, thinkplot, thinkstats2
import numpy as np
df = nsfg.ReadFemPreg()
df = df[np.isfinite(df['birthwgt_lb'])]
df = df[np.isfinite(df['agepreg'])]
sample = thinkstats2.SampleRows(df, 5000)
birth_weight, mother_age = sample.birthwgt_lb, sample.agepreg
birth_weight = thinkstats2.Jitter(birth_weight, 0.5)
thinkplot.Scatter(birth_weight, mother_age, alpha=0.2)
thinkplot.Show(xlabel='birth weight', ylabel='mother\'s age')
```

```python
birth_weight, mother_age = df.birthwgt_lb, df.agepreg
p = thinkstats2.Corr(birth_weight, mother_age)
print("Pearson's correlation is {0}".format(p))
```

```python
s = thinkstats2.SpearmanCorr(birth_weight, mother_age)
print("Spearman's correlation is {0}".format(s))
```

```python
print("Seems to be a pretty weak correlation even with outliers taken into account.")
```

####Answer:
For the Pearson r correlation, both variables should be normally distributed.  Other assumptions include linearity and homoscedasticity.  Linearity assumes a straight line relationship between each of the variables in the analysis and homoscedasticity assumes that data is normally distributed about the regression line.

Spearman rank correlation test does not make any assumptions about the distribution.
