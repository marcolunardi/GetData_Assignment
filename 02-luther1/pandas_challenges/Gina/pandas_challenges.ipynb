```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

movies = pd.read_csv('2013_movies.csv', header=0, index_col=6,parse_dates=True)
```

```python
#**Challenge 1**
#Plot domestic total gross over time


movies['DomesticTotalGross'].plot()
plt.title("Domestic Total Gross over time")
plt.xlabel("Time")
plt.ylabel("Domestic Total Gross")
```

```python
#**Challenge 2**
#Plot runtime vs domestic total gross

plt.scatter(movies['Runtime'], movies['DomesticTotalGross'])
plt.title('Runtime vs Domestic Total Gross')
plt.xlabel('Domestic Total Gross')
plt.ylabel('Runtime')
```

```python
#**Challenge 4**
#Make one figure with (N=the number of MPAA ratings there are) subplots, 
#and in each plot the release date vs the domestic total gross

g = movies[movies['Rating'] == 'G']
pg = movies[movies['Rating'] == 'PG']
pg13 = movies[movies['Rating'] == 'PG-13']
r = movies[movies['Rating'] == 'R']

fig, axes = plt.subplots(nrows=2, ncols=2)

g['DomesticTotalGross'].plot(ax=axes[0,0]); axes[0,0].set_title('G Rated')
pg['DomesticTotalGross'].plot(ax=axes[0,1]); axes[0,1].set_title('PG Rated')
pg13['DomesticTotalGross'].plot(ax=axes[1,0]); axes[1,0].set_title('PG-13 Rated')
r['DomesticTotalGross'].plot(ax=axes[1,1]); axes[1,1].set_title('R Rated')
fig.tight_layout()
```
